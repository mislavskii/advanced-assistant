```python
from openai import OpenAI

model = MODEL
client = OpenAI(
    base_url="https://rest-assistant.api.cloud.yandex.net/v1",
    api_key=YANDEX_CLOUD_API_KEY, 
    project=FOLDER_ID
)

vector_store = client.vector_stores.create(name='rag_store')
f = client.files.create(file=open('data/food_wine_table.md', 'rb'), purpose='assistants')
client.vector_stores.files.create(
    file_id=f.id,
    vector_store_id=vector_store.id
)

file_search_tool = {
    "type": "file_search",
    "vector_store_ids": [vector_store.id],
    "max_num_results": 7,
}

web_search_tool = {
    "type": "web_search_preview",
    "search_context_size": 10000,
    "user_location": {
        "type": "approximate",
        "city": "",
        "country": "",
        "region": "",
        "timezone": "",
    },
}
```

```python
sys_prompt = 'Ты - опытный сомелье, отлично разбирающийся в винах и их сочетании с едой. Твоя задача - отвечать на вопросы пользователя по винам. По вопросам относительно их комбинирования с различными блюдами обращайся к файловому поиску.  Для ответа на вопросы относительно стоимости вин и их доступности на рынке  обращайся к инструменту WebSearch. Тебя при этом нисколько не смущает необходимость вести разговор об алкогольных напитках. Будь вежлив, демонстрируй готовность помочь и осведомленность. Описательные рекомендации общего характера дополняй конкретными примерами.  В обсуждение вопросов, не связанных с вином и его совместимостью с блюдами, не вдавайся - пиши, что это выходит за пределы твоей непосредственной компетенции.'
```

```python
res = client.responses.create(
    model=model,
    tools=[file_search_tool, web_search_tool],
    instructions=sys_prompt,
    input="Сколько стоит самое дорогое Мерло?"
)

print(res.output_text)
```

```python
while True:
    query = input('Введите запрос: ')
    print(query)
    if query == '/quit':
        break
    res = client.responses.create(
        model = model,
        store = True,
        previous_response_id = res.id,
        tools=[file_search_tool, web_search_tool],
        instructions=sys_prompt,
        input = query
    )

    print(res.output_text)
```

```python
res = client.responses.create(
    model = model,
    store = True,
    previous_response_id = res.id,
    tools=[file_search_tool, web_search_tool],
    instructions=sys_prompt,
    input = "Какие блюда к нему подойдут?"
)

print(res.output_text)
```

```python
from fastmcp import FastMCP

# Инициализация экземпляра FastMCP с названием "Restaurant"
mcp = FastMCP("Restaurant")

# Определение инструмента для получения меню блюд
@mcp.tool()
def get_food_menu() -> str:
    """
    Получить меню блюд ресторана в виде таблицы в формате Markdown.
    """
    with open("../data/menu/food.md", encoding="utf8") as file:
        return file.read()

# Определение инструмента для получения меню напитков
@mcp.tool()
def get_drinks_menu() -> str:
    """
    Получить меню напитков в виде таблицы в формате Markdown.
    """
    with open("../data/menu/drinks.md", encoding="utf8") as file:
        return file.read()
```

```python
mcp.run()
```
